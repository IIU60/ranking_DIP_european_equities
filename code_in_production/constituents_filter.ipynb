```python
dates_dict = {}
current_list = og_list
missing_rics = []
for date in dates:
    if date in jl_df.Date.values:
        rows = jl_df.loc[jl_df.Date == date]
        for _, row in rows.T.items():
            try:
                if row.Change == 'Leaver':
                    current_list.remove(row['Constituent RIC'])
                else:
                    current_list.append(row['Constituent RIC'])
            except ValueError:
                missing_rics.append(row['Constituent RIC'])
    dates_dict[date] = current_list
```
